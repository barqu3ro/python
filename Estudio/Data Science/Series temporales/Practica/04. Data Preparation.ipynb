# Preparando datos en series temporales

Una técnica muy utilizada para trabajar con series temporales es el 'Moving Average Smoothing'

Esta técnica sirve muy bien tanto para análisis como para predicción.

Hay dos tipos:

 - Centered Moving Average: el valor en un instante es calculado por la media de las observaciones anteriores y posteriores.
 - Trailing Moving Average: el valor en un instante es calculado por la media de las observaciones anteriores.
 
Como podemos ver, la primera sirve para procesar los datos y la segunda para hacer forecasting.

En este notebook hablaremos de Centered Moving Average.

```python
import warnings
warnings.filterwarnings(action="ignore")

%matplotlib inline
from matplotlib.pylab import plt
from matplotlib.pyplot import figure

import pandas as pd
import numpy as np
```

##### Data preparation

```python
df = pd.DataFrame(np.random.randn(600, 1), index = pd.date_range('15/1/2021', freq = 'D', periods = 600), columns = ['Value'])

df.head()
```

```python
WINDOW_SIZE = 20

r = df.rolling(window = WINDOW_SIZE, center=True)

figure(figsize=(18, 6))
df['Value'].plot(label='Value');
r.mean()['Value'].plot(label='Mean (WINDOW_SIZE: '+str(WINDOW_SIZE) + ')', color='green');
plt.legend();
```

```python
WINDOW_SIZE = 5

r = df.rolling(window = WINDOW_SIZE, center=True)

figure(figsize=(18, 6))
df['Value'].plot(label='Value');
r.mean()['Value'].plot(label='Mean (WINDOW_SIZE: '+str(WINDOW_SIZE) + ')', color='green');
plt.legend();
```

```python
WINDOW_SIZE = 100

r = df.rolling(window = WINDOW_SIZE, center=True)

figure(figsize=(18, 6))
df['Value'].plot(label='Value');
r.mean()['Value'].plot(label='Mean (WINDOW_SIZE: '+str(WINDOW_SIZE) + ')', color='green');
plt.legend();
```

Vemos cómo podemos suavizar los datos quitando el ruido y obtener una media a lo largo de la serie temporal que nos indique un comportamiento que nos interese más.

Otras funciones agregadas también se pueden aplicar, como son el máximo, mínimo... incluso funciones propias usando lambda.

```python
WINDOW_SIZE = 20

r = df.rolling(window = WINDOW_SIZE, center=True)

figure(figsize=(18, 6))
df['Value'].plot(label='Value');
r.max()['Value'].plot(label='Max (WINDOW_SIZE: '+str(WINDOW_SIZE) + ')', color='green');
plt.legend();
```

```python
WINDOW_SIZE = 20

r = df.rolling(window = WINDOW_SIZE, center=True)

figure(figsize=(18, 6))
df['Value'].plot(label='Value');
r.min()['Value'].plot(label='Min (WINDOW_SIZE: '+str(WINDOW_SIZE) + ')', color='green');
plt.legend();
```

```python
WINDOW_SIZE = 20

r = df.rolling(window = WINDOW_SIZE, center=True)['Value'].apply(lambda x: 10)

figure(figsize=(18, 6))
df['Value'].plot(label='Value');
r.plot(label='Lambda Function (WINDOW_SIZE: '+str(WINDOW_SIZE) + ')', color='green');
plt.legend();
```
