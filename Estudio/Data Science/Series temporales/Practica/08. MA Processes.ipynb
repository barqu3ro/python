# Procesos de medias móviles (MA)

Los procesos de medias móviles son procesos estacionarios que nos dicen que en una serie temporal el valor en un instante dado depende de la componente aleatoria ruidosa en los q instantes anteriores.

```python
import warnings
warnings.filterwarnings(action="ignore")

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.arima_model import ARIMA

from sklearn.metrics import mean_squared_error
from math import sqrt
```

### Simulamos un proceso MA de orden 1

```python
# Simulamos un proceso MA de orden 1
ma1 = np.array([1, 0.6])
q = len(ma1)-1
N_SAMPLES = 1000

simulated_MA1 = ArmaProcess(ma=ma1).generate_sample(nsample=N_SAMPLES)

# Visualizamos el proceso MA de orden 1
plt.figure(figsize=[10, 6])
plt.plot(simulated_MA1)
plt.title("Simulated MA(1) Process")
plt.show()
```

```python
# Visualizamos las autocorrelaciones
fig, ax = plt.subplots(1,2,figsize=(10,5))
plot_acf(simulated_MA1, lags=20, ax=ax[0])
plot_pacf(simulated_MA1, lags=20, ax=ax[1])
plt.show()
```

```python
# Vemos en la autocorrelacion el orden interesante
acf_coef = acf(simulated_MA1)
print(acf_coef)
```

```python
# Entrenamos el modelo
model = ARIMA(simulated_MA1, order=(0, 0, q))
results_MA1 = model.fit(disp=-1)

# Visualizamos las predicciones vs la serie original
plt.figure(figsize=[10, 6])
plt.plot(simulated_MA1)
plt.plot(results_MA1.fittedvalues, '-r')

# Métrica
rmse = sqrt(mean_squared_error(simulated_MA1, results_MA1.fittedvalues))
print("RMSE:", np.round(rmse, 3))
```
