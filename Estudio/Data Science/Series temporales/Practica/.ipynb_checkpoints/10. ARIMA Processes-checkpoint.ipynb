# Procesos ARIMA

Son los procesos no estacionarios a los que se le aplica differencing para pasarlos a estacionario y entonces pasarles un proceso ARMA.

```python
import  scipy.signal.signaltools

def _centered(arr, newsize):
    # Return the center newsize portion of the array.
    newsize = np.asarray(newsize)
    currsize = np.array(arr.shape)
    startind = (currsize - newsize) // 2
    endind = startind + newsize
    myslice = [slice(startind[k], endind[k]) for k in range(len(endind))]
    return arr[tuple(myslice)]

scipy.signal.signaltools._centered = _centered
```

```python
import warnings
warnings.filterwarnings(action="ignore")

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd
%matplotlib inline

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_model import ARIMA

from sklearn.metrics import mean_squared_error
from math import sqrt
```

## Shampoo Sales Dataset

```python
# Funciones
def parser(x):
    return pd.datetime.strptime('190'+x, '%Y-%m')
```

```python
# Cargamos el dataset
real = pd.read_csv('data/Shampoo-Sales.csv', header=0, parse_dates=['Month'], index_col=0,  date_parser=parser)
```

```python
# Visualizamos el dataset
ax = real.plot(figsize=(18,6), title='Venta de shampoo');
ax.set_xlabel("Fecha");
ax.set_ylabel("Ventas");
```

```python
def splitSet(df, TEST_SIZE):
    
    # Train set
    train_full = df[:(int(len(df)*(1-TEST_SIZE)))]
    train = train_full[:(int(len(train_full)*(1-TEST_SIZE)))]

    # Validation set
    valid = train_full[(int(len(train_full)*(1-TEST_SIZE))):]

    # Test set
    test = df[(int(len(df)*(1-TEST_SIZE))):]

    return train, valid, test
```

```python
TEST_SIZE = 0.24
train, valid, test = splitSet(real, TEST_SIZE)

# Trabajamos slo con train y test
train = pd.concat([train, valid], axis=0)

# Visualizacin
ax = real.plot(figsize=(18,6), title='Venta medicamentos');

init__limit = real.index[0]
train_limit = real.index[len(train)-1]
test__limit = real.index[len(train)+len(test)-1]

ax.axvline(train_limit, linestyle='--', color='blue')
ax.axvline(test__limit, linestyle='--', color='red')

ax.axvspan(init__limit, train_limit, alpha=0.1, color='blue')
ax.axvspan(train_limit, test__limit, alpha=0.1, color='red')

ax.text(real.index[int((len(train)-1)*0.5)], 610, 'Train set', fontsize=18)
ax.text(real.index[int(len(train)+(len(test)-1)*0.3)], 610, 'Test set', fontsize=18)

ax.grid(linestyle='--')

ax.set_xlabel("Fecha");
ax.set_ylabel("Ventas");
```

```python
shampooSalesMetrics = pd.DataFrame(columns=['Model', 'RMSE'])
```

Vamos a comparar cmo seran los resultados si nos hubieramos enfrentado a una serie temporal con el mtodo ARMA y con ARIMA. El avance que supuso ejecutar el differencing para obtener resultados fue muy relevante.

## ARMA

```python
# Model
MODEL = 'ARMA Process'

# Coefficients
p=1
q=1

# Training model
model = ARIMA(train, order=(p, 0, q))
fitted_model = model.fit(disp=-1)
fitted_train = fitted_model.fittedvalues

# Get Test Set Predictions
predict_test = fitted_model.predict(test.index[0], test.index[len(test)-1])

# Get RMSE from Test Set
real_values = test['Sales']
pred_values = predict_test
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = fitted_model.predict(real.index[0], real.index[len(real)-1])

# Preparing data to plot
real_plot = real['Sales']
pred_plot = predictions

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
shampooSalesMetrics = shampooSalesMetrics.append(metric_data, ignore_index=True)
```

## ARIMA

### Box-Jenkins - Identificacin del modelo

```python
# Descomponemos la serie para entender cmo es la estructura
additive = seasonal_decompose(real, model='additive', freq=12)
with plt.rc_context():
    plt.rc("figure", figsize=(18,6))
    additive.plot()
    plt.show()
```

```python
# Vemos las autocorrelaciones para comprobar los ordenes y analizar la estructura
fig, ax = plt.subplots(1,2,figsize=(18,6))
plot_acf(real, lags=17, ax=ax[0])
plot_pacf(real, lags=17, ax=ax[1])
plt.show()
```

### Box-Jenkins - Estimacin de parmetros y evaluacin en test set

```python
# Model
MODEL = 'ARIMA Process (2, 1, 1)'

# Coefficients
p=2
d=1
q=1

# Training model
model = ARIMA(train, order=(p, d, q))
fitted_model = model.fit(disp=-1)
fitted_train = fitted_model.fittedvalues

# Get Test Set Predictions
predict_test = fitted_model.predict(test.index[0], test.index[len(test)-1])

# Reverting from differencing
fitted_train = fitted_train + train.shift(periods=d)['Sales']
predict_test = predict_test + test.shift(periods=d)['Sales']

# Get RMSE from Test Set
real_values = test['Sales'][d:]
pred_values = predict_test[d:]
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = fitted_model.predict(real[d:].index[0], real[d:].index[len(real[d:])-1])

# Reverting from differencing
predictions = predictions + real[d:].shift(periods=d)['Sales']

# Preparing data to plot
real_plot = real['Sales'][d:]
pred_plot = predictions

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
shampooSalesMetrics = shampooSalesMetrics.append(metric_data, ignore_index=True)
```

Anlisis

 - Tiene tendencia, luego seguramente necesitemos hacer como mnimo un differencing: <strong>d=1</strong>
 - Hay una correlacin muy interesante en los 12 primeros lags (ACF), es posible que sea por los 5 primeros ya que a partir del 5 vemos cmo se repite: <strong>p=5</strong>
 - No tenemos muy claro qu orden tiene MA, lo dejamos a 0 y vemos ms adelante (un buen primer paso es empezar con AR y luego ir iterando): <strong>q=0</strong>

```python
# Model
MODEL = 'ARIMA Process (5, 1, 0)'

# Coefficients
p=5
d=1
q=0

# Training model
model = ARIMA(train, order=(p, d, q))
fitted_model = model.fit(disp=-1)
fitted_train = fitted_model.fittedvalues

# Get Test Set Predictions
predict_test = fitted_model.predict(test.index[0], test.index[len(test)-1])

# Reverting from differencing
fitted_train = fitted_train + train.shift(periods=d)['Sales']
predict_test = predict_test + test.shift(periods=d)['Sales']

# Get RMSE from Test Set
real_values = test['Sales'][d:]
pred_values = predict_test[d:]
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = fitted_model.predict(real[d:].index[0], real[d:].index[len(real[d:])-1])

# Reverting from differencing
predictions = predictions +  + real[d:].shift(periods=d)['Sales']

# Preparing data to plot
real_plot = real['Sales'][d:]
pred_plot = predictions

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
shampooSalesMetrics = shampooSalesMetrics.append(metric_data, ignore_index=True)
```
