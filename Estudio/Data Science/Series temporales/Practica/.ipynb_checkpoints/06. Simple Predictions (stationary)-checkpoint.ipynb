# Prediciones simples en series temporales (estacionarias)

En este notebook vamos a retomar lo hecho en el notebook '05. Simple Predictions - TS' pero ahora si vamos a transformar las series temporales en estacionarias, para as√≠ ver el efecto que tiene tratar de predecir una serie temporal siendo estacionaria.

```python
import warnings
warnings.filterwarnings(action="ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from sklearn.metrics import mean_squared_error
from math import sqrt
```

## Creando una TS manualmente (la misma que en el ejercicio anterior)

Queremos que tenga tendencia, estacionalidad y ruido

(Funciones obtenidas de deeplearning.ai)

```python
def trend(time, slope=0):
    return slope * time

def seasonal_pattern(season_time):
    """Just an arbitrary pattern, you can change it if you wish"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repeats the same pattern at each period"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)

def noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level

time = np.arange(4 * 365 + 1, dtype="float32") # 4 a√±os
baseline = 10
series = trend(time, 0.1)  
baseline = 10
amplitude = 40
slope = 0.05
noise_level = 5

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))
fig.suptitle('Building a custom TS')

# Base + Trend
series = baseline + trend(time, slope)
ax1.plot(time[0:], series[0:])

# Base + Trend + Seasonality
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude)
ax2.plot(time[0:], series[0:])

# Base + Trend + Seasonality + Noise
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude) + noise(time, noise_level, seed=42)
ax3.plot(time[0:], series[0:]);
```

Recordemos que el set de datos se pod√≠a dividir de varias formas para hacer forecasting.

Una de ellas era seleccionar un set de train suficientemente grande y un set de validaci√≥n que corresponda al √∫ltimo bloque, es decir, lo m√°s reciente. Este set de validaci√≥n deber√° ser un m√∫ltiplo estacional.

```python
# Creamos el dataframe con la serie reci√©n creada

rng = pd.date_range('2017 Jul 02', periods = 4*365+1, freq = 'd')

real = pd.DataFrame(series, index=rng, columns=['Value'])

ax = real.plot(figsize=(18,6), title='Custom Time Series');
ax.set_xlabel("Days");
ax.set_ylabel("Value");
```

```python
def splitSet(df, TEST_SIZE):
    
    # Train set
    train_full = df[:(int(len(df)*(1-TEST_SIZE)))]
    train = train_full[:(int(len(train_full)*(1-TEST_SIZE)))]

    # Validation set
    valid = train_full[(int(len(train_full)*(1-TEST_SIZE))):]

    # Test set
    test = df[(int(len(df)*(1-TEST_SIZE))):]

    return train, valid, test
```

```python
TEST_SIZE = 0.25
train, valid, test = splitSet(real, TEST_SIZE)

# Trabajamos s√≥lo con train y test
train = pd.concat([train, valid], axis=0)
```

```python
ax = real.plot(figsize=(18,6), title='Custom Time Series');

init__limit = real.index[0]
train_limit = real.index[len(train)-1]
test__limit = real.index[len(train)+len(test)-1]

ax.axvline(train_limit, linestyle='--', color='blue')
ax.axvline(test__limit, linestyle='--', color='red')

ax.axvspan(init__limit, train_limit, alpha=0.1, color='blue')
ax.axvspan(train_limit, test__limit, alpha=0.1, color='red')

ax.text(real.index[int((len(train)-1)*0.5)], 110, 'Train set', fontsize=18)
ax.text(real.index[int(len(train)+(len(test)-1)*0.3)], 110, 'Test set', fontsize=18)

ax.grid(linestyle='--')

ax.set_xlabel("Days");
ax.set_ylabel("Value");
```

```python
customTSMetrics = pd.DataFrame(columns=['Model', 'RMSE'])
```

### Naive Forecast

```python
# Model
MODEL = 'Naive Method'

# Training model
fitted_train = train.shift(periods=1)

# Get Test Set Predictions
predict_test = test.shift(periods=1)

# Get RMSE from Test Set
real_values = test['Value'][1:]
pred_values = predict_test['Value'][1:]
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = real.shift(periods=1)

# Preparing data to plot
real_plot = real['Value']
pred_plot = predictions['Value']

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
customTSMetrics = pd.concat([customTSMetrics, pd.DataFrame([metric_data])], ignore_index=True)
```

### Trailing Moving Average

```python
# Model
MODEL = 'Trailing Moving Average'

# Training model
WINDOW_SIZE = 30
fitted_train = train.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Get Test Set Predictions
predict_test = test.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Get RMSE from Test Set
real_values = test['Value'][WINDOW_SIZE:]
pred_values = predict_test['Value'][WINDOW_SIZE:]
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = real.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Preparing data to plot
real_plot = real['Value']
pred_plot = predictions['Value']

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
customTSMetrics = pd.concat([customTSMetrics, pd.DataFrame([metric_data])], ignore_index=True)
```

Vamos a probar a hacer differencing.
Recordemos: esta t√©cnica consiste en restar al valor en el instante actual un valor separado un n√∫mero de instantes.

En nuestro caso hemos creado una serie estacional, por lo que podemos hacer esa differencing con el valor 365.

### Haci√©ndola estacionaria

#### Trabajando con Differencing - Order 1

```python
ORDER_DIFF = 1

real_diff_1 = real - real.shift(periods=ORDER_DIFF)
real_diff_1.head(10)
```

```python
ax = real_diff_1.plot(figsize=(18,6), title='Custom Time Series');
ax.set_xlabel("Days");
ax.set_ylabel("Value");
```

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(real_diff_1.dropna())
print("The p-value: ", adf_test[1])
```

Genial, ¬°es estacionaria!

Ejecutemos ahora Trailing Moving Average a ver si ahora funciona mejor.

```python
# Model
MODEL = 'Trailing Moving Average - Diff 1'

# Applying differencing
real_diff_1 = real - real.shift(periods=ORDER_DIFF)
train_diff_1 = train - train.shift(periods=ORDER_DIFF)
test_diff_1 = test - test.shift(periods=ORDER_DIFF)

# Training model
WINDOW_SIZE = 30
fitted_train = train_diff_1.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Get Test Set Predictions
predict_test = test_diff_1.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Reverting from differencing
fitted_train = fitted_train + train.shift(periods=ORDER_DIFF)
predict_test = predict_test + test.shift(periods=ORDER_DIFF)

# Get RMSE from Test Set
real_values = test['Value'][WINDOW_SIZE+ORDER_DIFF:]
pred_values = predict_test['Value'][WINDOW_SIZE+ORDER_DIFF:]
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = real_diff_1.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Reverting from differencing
predictions = predictions +  + real.shift(periods=ORDER_DIFF)

# Preparing data to plot
real_plot = real['Value']
pred_plot = predictions['Value']

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
customTSMetrics = pd.concat([customTSMetrics, pd.DataFrame([metric_data])], ignore_index=True)
```

```python
customTSMetrics.sort_values(by=['RMSE'], ascending=True)
```

Lo hemos mejorado mucho respecto al valor anterior en Trailing Moving Average y susceptiblemente mejor al anterior Naive Forecast, luego ya hemos comprobado que si la serie es estacionaria la predicci√≥n ser√° mejor.

Si siguieramos ejecutando medias m√≥viles, seguramente podr√≠amos quitar el ruido y mejorar la predicci√≥n.

#### Trabajando con el logaritmo y differencing order 1

```python
# Logaritmo
real_log = np.log(real)

# Differencing
ORDER_DIFF = 1

real_diff_1 = real_log - real_log.shift(periods=ORDER_DIFF)
real_diff_1.head(10)
```

```python
# Visualization
ax = real_diff_1.plot(figsize=(18,6), title='Pasajeros de avi√≥n (differencing order '+str(ORDER_DIFF)+' and Log)');
ax.set_xlabel("Fecha");
ax.set_ylabel("Pasajeros (differencing and log)");
```

```python
# Model
MODEL = 'Trailing Moving Average (Log + Differencing Order 1)'

# Applying log
real_log = np.log(real)
train_log = np.log(train)
test_log = np.log(test)

# Applying differencing
real_diff_1 = real_log - real_log.shift(periods=ORDER_DIFF)
train_diff_1 = train_log - train_log.shift(periods=ORDER_DIFF)
test_diff_1 = test_log - test_log.shift(periods=ORDER_DIFF)

# Training model
WINDOW_SIZE = 5
fitted_train = train_diff_1.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Get Test Set Predictions
predict_test = test_diff_1.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Reverting from differencing
fitted_train = fitted_train + train_log.shift(periods=ORDER_DIFF)
predict_test = predict_test + test_log.shift(periods=ORDER_DIFF)

# Get normal scale
fitted_train = np.exp(fitted_train)
predict_test = np.exp(predict_test)

# Get RMSE from Test Set
real_values = test['Passengers'][WINDOW_SIZE+ORDER_DIFF:]
pred_values = predict_test['Passengers'][WINDOW_SIZE+ORDER_DIFF:]
rmse = sqrt(mean_squared_error(real_values, pred_values))

# Get predictions from whole set
predictions = real_diff_1.rolling(window = WINDOW_SIZE, center=False, closed='left').mean()

# Reverting from differencing
predictions = predictions + real_log.shift(periods=ORDER_DIFF)

# Get normal scale
predictions = np.exp(predictions)

# Preparing data to plot
real_plot = real['Passengers']
pred_plot = predictions['Passengers']

# Visualization
figure(figsize=(18, 6))

real_plot.plot(label='Real Value')
pred_plot.plot(label='Predictions', color='red')
plt.suptitle(MODEL + ' - RMSE: ' + str(np.round(rmse, 3)), fontsize=20)
plt.legend();

plt.axvline(train_limit, linestyle='--', color='blue');
plt.axvline(test__limit, linestyle='--', color='red');

plt.axvspan(init__limit, train_limit, alpha=0.1, color='blue');
plt.axvspan(train_limit, test__limit, alpha=0.1, color='red');

plt.grid(linestyle='--')

metric_data = {'Model':MODEL, 'RMSE':np.round(rmse, 3)}
airPassMetrics = pd.concat([airPassMetrics, pd.DataFrame([metric_data])], ignore_index=True)
```
