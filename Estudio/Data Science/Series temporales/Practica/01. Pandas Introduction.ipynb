### Creando Series en Pandas

```python
# Podemos trabajar con Series, que es como un dataframe pero de 1 dimension
s = pd.Series([1,3,5,np.nan,6,8])
s.index = ['a', 'b', 'c', 'd', 'e', 'f']
s.head()
```

#### Visualizando los datos

```python
s.plot();
```

#### Analizando los datos

```python
s.isnull()
```

```python
s[s.index > 'c']
```

```python
s[s.isnull() == False]
```

### Abriendo archivos CSV

```python
# Dataset
# https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data?select=GlobalTemperatures.csv

df = pd.read_csv('data/GlobalTemperatures.csv')
df.head()
```

```python
# Pasando la fecha como index
df = pd.read_csv('data/GlobalTemperatures.csv', parse_dates=['dt'], index_col='dt')
df.head()
```

#### Visualizando los datos

```python
# Visualizando los datos
df.plot(figsize=(16,5), title='Temperatura Global');
```

```python
# Seleccionando una columna
df_avgtemp = df['LandAverageTemperature']
```

```python
df_avgtemp.head()
```

```python
df_avgtemp.plot(figsize=(16,5), title='Media Temperatura Global');
```

```python
# Poniendo etiquetas a los ejes
ax = df_avgtemp.plot(figsize=(16,5), title='Media Temperatura Global');
ax.set_xlabel("Fecha");
ax.set_ylabel("Temperatura");
```

```python
df_zoom = df_avgtemp[(df_avgtemp.index.year >= 1820) & (df_avgtemp.index.year <= 1840)]

df_zoom.plot(figsize=(16,5), title='Temperatura Global (1820-1840)');
ax.set_xlabel("Fecha");
ax.set_ylabel("Temperatura");
```
