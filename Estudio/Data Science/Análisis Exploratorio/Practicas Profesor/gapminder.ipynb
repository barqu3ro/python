2.- Cargar los datos con Pandas e imprimir las cinco primeras filas del DataFrame gapminder. Describir los datos y comentar los resultados.

```python
import os
import pandas as pd 
csv_path = os.path.join(os.getcwd(), 'data', 'gapminder.csv')


#print
datos=pd.read_csv(csv_path) 
datos.head()
datos.describe()
```

```python
sns.histplot(data=datos, x="lifeExp")
```

```python
#Para verlo como una linea continua
sns.kdeplot(datos['lifeExp'], bw=0.5)
```

```python
# como ambas cosas
sns.distplot(datos['lifeExp'])
```

```python
sns.boxplot(x="lifeExp", data=datos)
```

```python
sns.scatterplot(data=datos, x="lifeExp", y="gdpPercap", hue="continent")
```
