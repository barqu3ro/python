```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data
data = pd.read_csv('data/customer-entrenamiento.csv')
# Datos de cabecera del archivo.
data.head()
```

```python
data.describe ()
```

```python
#import matplotlib.pyplot as plt

plt.scatter(data['AGE'], data['NETPRICE_PRO14_AMT'])
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('Scatter Plot: Age vs NETPRICE_PRO14_AMT')
plt.show()
```

```python
#Intento de clusterizar

from sklearn.cluster import KMeans

# Create an instance of the KMeans class
kmeans = KMeans(n_clusters=3)

# Fit the data to the KMeans model
kmeans.fit(data[['AGE', 'NETPRICE_PRO14_AMT']])

# Get the cluster labels
labels = kmeans.labels_

# Add the cluster labels to the dataframe
data['Cluster'] = labels

# Display the updated dataframe
data.head()
```
